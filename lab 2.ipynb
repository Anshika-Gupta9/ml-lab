```python
import numpy as np

lst = [10, 20, 30, 40, 50]
arr = np.array(lst)

print("Array:", arr)
print("Shape:", arr.shape)
print("Data Type:", arr.dtype)
```

```python
import numpy as np

zeros_matrix = np.zeros((3, 3))
ones_matrix = np.ones((2, 5))

print("3x3 Zero Matrix:\n", zeros_matrix)
print("Shape:", zeros_matrix.shape)

print("\n2x5 Ones Matrix:\n", ones_matrix)
print("Shape:", ones_matrix.shape)
```

```python
import numpy as np

arr = np.arange(1, 13)
reshaped_arr = arr.reshape(3, 4)

print("Original Array:", arr)
print("Original Shape:", arr.shape)

print("\nReshaped Array:\n", reshaped_arr)
print("Reshaped Shape:", reshaped_arr.shape)
```

```python
import numpy as np

arr = np.random.randint(1, 100, 10)

print("Array:", arr)
print("Minimum:", arr.min())
print("Maximum:", arr.max())
print("Mean:", arr.mean())
print("Sum:", arr.sum())
```

```python
import pandas as pd

data = {
    "Student_Name": ["Arshi", "Aman", "Riya", "Karan"],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print(df)
print("\nData Types:\n", df.dtypes)
```

```python
import pandas as pd

data = {
    "Name": ["Arshi", "Aman", "Riya", "Karan", "Neha", "Rahul"],
    "Marks": [85, 90, 78, 88, 92, 80]
}

df = pd.DataFrame(data)
df.to_csv("data.csv", index=False)

print("CSV file created successfully")
```
