```python
import numpy as np
a = np.array([1, 2, 3])
print(a)
```

```python
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print(b)
```

```python
c = np.array([1, 2, 3], dtype=float)
print(c)
```

```python
d = np.array([[1, 2],
              [3, 4],
              [5, 6],
              [7, 8]])
print(d)
```

```python
e = np.arange(10)
print(e)
```

```python
f = np.arange(16).reshape(2, 2, 2, 2)
print(f)
```

```python
g = np.ones((2, 4))
print(g)
```

```python
h = np.random.random((3, 4))
print(h)

```

```python
i = np.identity(3)
print(i)

```
